```python
saldo = 2000.00

while True:
    print("\n--- CAIXA ELETRÔNICO ---")
    print("1 - Verificar Saldo")
    print("2 - Depositar")
    print("3 - Sacar")
    print("4 - Sair")
    opcao = input("Escolha uma opção (1-4): ")
    
    if opcao == "1":
        print(f"Seu saldo é R$ {saldo:.2f}")
    elif saldo and opcao == "2":
        valorDeDeposito = float(input("Digite o valor que gostaria de depositar:"))
        if valorDeDeposito > 0:
            saldo += valorDeDeposito
            print(f"Seu saldo atual é de R$ {saldo:.2f}.")
        else:
            print("Digite um valor que seja maior que R$ 1.")
    elif saldo >= 0 and opcao == "3":
        valorDeSaque = float(input("Digite o valor que gostaria de sacar:"))
        if valorDeSaque > 0 and saldo >= valorDeSaque:
            saldo -= valorDeSaque
            print(f"Seu saldo atual é de R$ {saldo:.2f}.")
        else:
            print("Saldo insuficiente para saque. Tente novamente!")
    elif opcao == "4":
        print("Obrigado por utilizar nossos serviços. Volte sempre!")
        break
    else:
        print("Opção inválida. Digite uma opção entre 1-4.")
```
