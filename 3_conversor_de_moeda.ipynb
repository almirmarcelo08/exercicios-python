```python
dolar = 5

while True:
    print("\n--- BAZAR MÁGICO DE MOEDAS ---")
    opcao = input("1 - Dólar para Real\n2 - Real para Dólar\n3 - Sair\nEscolha: ")

    if opcao == "3":
        break

    if opcao in ["1", "2"]:
        valor = float(input("Quanto você quer trocar? "))
        
        if opcao == "1":
            resultado = valor * dolar
            print(f"Conversão: R${resultado:.2f}")
        else:
            resultado = valor / dolar
            print(f"Conversão: ${resultado:.2f}")            
        
        continuar = input("Deseja continuar? (Sim/Não): ")
        if continuar.lower() == "não":
            break
    else:
        print("Opção inválida!")

print("Programa encerrado. Volte sempre!")
```
